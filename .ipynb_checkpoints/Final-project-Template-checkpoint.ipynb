```python
## null value analysis, null ratio in row and in column
# NOTE: check which value represents NULL value

data.replace('na', np.nan, inplace=True)
```

```python
col_stats = data.isnull().sum(axis=0) / data.shape[0]
row_stats = data.isnull().sum(axis=1) / data.shape[1]
```

```python
col_stats
```

```python
g = sns.distplot(col_stats.values)
```

```python
row_stats.max()
```

```python
g = sns.distplot(row_stats.values)
```

```python
cols_to_drop = col_stats.index[col_stats >  0.2]
data_dropped = data.drop(cols_to_drop, axis=1)
```

```python
rows_to_drop = row_stats.index[row_stats >  0.4]
data_dropped = data_dropped.drop(rows_to_drop, axis=0)
```

```python
data_dropped.shape
```

```python
data_dropped = data_dropped.drop(columns=['class'])
```

```python
data_dropped = data_dropped.astype('float')
```

```python
data_dropped.head()
```

```python
col_mean = data_dropped.mean(axis=0)
col_mean.shape
```

```python
data_dropped = data_dropped.fillna(data_dropped.mean())
```
